```python
import time
import random

# Função para obter o dígito máximo
def get_max(arr):
    return max(arr)

# Função para ordenar usando um dígito específico (counting sort adaptado)
def counting_sort(arr, exp):
    n = len(arr)
    output = [0] * n
    count = [0] * 10

    # Contar as ocorrências de cada dígito
    for i in range(n):
        index = arr[i] // exp
        count[index % 10] += 1

    # Modificar count[i] para conter a posição real de cada dígito
    for i in range(1, 10):
        count[i] += count[i - 1]

    # Construir o array de saída
    i = n - 1
    while i >= 0:
        index = arr[i] // exp
        output[count[index % 10] - 1] = arr[i]
        count[index % 10] -= 1
        i -= 1

    # Copiar o array de saída para arr
    for i in range(len(arr)):
        arr[i] = output[i]

# Função principal do Radix Sort
def radix_sort(arr):
    max_val = get_max(arr)
    exp = 1
    while max_val // exp > 0:
        counting_sort(arr, exp)
        exp *= 10

# Função para testar o tempo de execução para diferentes tamanhos
def test_radix_sort():
    sizes = [100, 1000, 10000, 100000, 1000000, 10000000]
    results = []

    for size in sizes:
        arr = [random.randint(0, size) for _ in range(size)]
        start_time = time.time()
        radix_sort(arr)
        end_time = time.time()
        results.append((size, end_time - start_time))

    print("Tamanho do Array | Tempo de Execução (s)")
    for size, duration in results:
        print(f"{size:<15} | {duration:.6f}")

test_radix_sort()
```
